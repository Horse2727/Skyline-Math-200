````python
#Manage data for variable 1
df['H4WS1'].value_counts()
````

````python
df['Siblings']= df['H4WS1']
df['Siblings']= pd.to_numeric(df['Siblings'],errors= 'coerce')
df['Siblings']= df['Siblings'].loc[df['Siblings'] < 96]
df['Siblings']= df['Siblings'].loc[df['Siblings'] != ' ']
df['Siblings'].value_counts()
````

````python
df['Siblings'].describe()
````

````python
sns.boxplot(x = 'Siblings', data =df)
````

````python
sns.histplot(x = 'Siblings', data =df, stat='percent', bins = 15,kde = True)
````

````python
#Manage data for variable 1
df['H4KK15A'].value_counts()
````

````python
df['Happy_being_parent'] = df['H4KK15A']
df['H4KK15A'].value_counts()
````

````python
df['Happy_being_parent']= df['Happy_being_parent'].loc[df['Happy_being_parent'] != ' ']
df['Happy_being_parent']= df['Happy_being_parent'].loc[df['Happy_being_parent'] != '96']
df['Happy_being_parent']= df['Happy_being_parent'].loc[df['Happy_being_parent'] != '97']
df['Happy_being_parent']= df['Happy_being_parent'].loc[df['Happy_being_parent'] != '98']
df['Happy_being_parent'].value_counts()
````

````python
df['Happy_being_parent'].replace(['1','2','3','4','5'], ['Strongly Agree','Agree','Neither Agree nor Disagree','Disagree','Strongly Disagree'],inplace=True)
df['Happy_being_parent'].value_counts()
````

````python
#Graphs and tables or statistics
plt.title("I am happy in my role as parent.")
df['Happy_being_parent'].value_counts(normalize = True).plot(kind = 'bar')
````
